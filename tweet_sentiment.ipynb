```python
import pandas as pd
import datetime

df_tweets_original = pd.read_csv(r'./tweet_dataset/Tweet.csv')
df_company_tweet = pd.read_csv(r'./tweet_dataset/Company_Tweet.csv')

df_tweets = pd.merge(df_tweets_original, df_company_tweet, on='tweet_id', how='left')

df_tweets['post_date'] =  pd.to_datetime(df_tweets['post_date'], unit='s').dt.floor('d')
df_tweets = df_tweets.loc[df_tweets['like_num'] > 2]
df_tweets_aapl = df_tweets.loc[df_tweets['ticker_symbol'] == 'AAPL']
df_tweets_aapl
```
