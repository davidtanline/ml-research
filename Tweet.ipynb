```python
import pandas as pd
tweets = pd.read_csv('tweet_data/Tweet.csv')
tweets = tweets[tweets['like_num'] > 2] #more than 2 likes
tweets['post_date'] = pd.to_datetime(tweets['post_date'],unit='s') #change epoch time 
```

```python
tweets_simplify = tweets.copy()
tweets_simplify
```

```python
tweets_simplify.drop('writer', inplace=True, axis=1)
tweets_simplify.drop('comment_num', inplace=True, axis=1)
tweets_simplify.drop('retweet_num', inplace=True, axis=1)
#remove tweet_id, writer, comment_num, retweet_num
```

```python
tweets_simplify
```

```python
tweets_simplify['post_date'] = tweets_simplify['post_date'] + pd.DateOffset(days=7) #offset date by 1 week
tweets_simplify = tweets_simplify[tweets_simplify['post_date'] < '2015-01-09'] #filter post_date
tweets_simplify
```

```python
company_tweet = pd.read_csv('tweet_data/Company_Tweet.csv')
company_tweet
```

```python
tweets_simplify = pd.merge(tweets_simplify, company_tweet, left_on='tweet_id', right_on='tweet_id')
tweets_simplify['post_date'] = tweets_simplify['post_date'].dt.normalize() #get rid of time 
```

```python
tweets_simplify
```

```python
body = [str(i) for i in tweets_simplify['body'].values]
ticker = [str(i) for i in tweets_simplify['ticker_symbol'].values]
```

```python
price = pd.read_csv('stock_price/CompanyValues.csv')
price['change'] = (price['close_value'] - price['open_value']).round(2) #calculate weekly change
price = price[price['day_date'] < '2015-01-09']
price = price[price['day_date'] > '2015-01-07']
price
```

```python
price['day_date'] = price['day_date'].astype('datetime64[ns]')
tweets_simplify = pd.merge(tweets_simplify, price, left_on=['post_date', 'ticker_symbol'], right_on=['day_date', 'ticker_symbol'], how = 'left')
```

```python
tweets_simplify.drop('close_value', inplace=True, axis=1)
tweets_simplify.drop('volume', inplace=True, axis=1)
tweets_simplify.drop('open_value', inplace=True, axis=1)
tweets_simplify.drop('high_value', inplace=True, axis=1)
tweets_simplify.drop('low_value', inplace=True, axis=1)
```

```python
tweets_simplify['post_date'] = tweets_simplify['post_date']+ pd.DateOffset(days=-7) #offset date by 1 week
apple_tweets = tweets_simplify.loc[tweets_simplify['ticker_symbol'] == 'AAPL']
apple_tweets
```
