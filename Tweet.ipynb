```python
import pandas as pd
tweets = pd.read_csv('tweet_data/Tweet.csv')
tweets = tweets[tweets['like_num'] > 2] #more than 2 likes
tweets['post_date'] = pd.to_datetime(tweets['post_date'],unit='s') #change epoch time 
```

```python
tweets_simplify = tweets.copy()
tweets_simplify
```

```python
tweets_simplify.drop('writer', inplace=True, axis=1)
tweets_simplify.drop('comment_num', inplace=True, axis=1)
tweets_simplify.drop('retweet_num', inplace=True, axis=1)
#remove tweet_id, writer, comment_num, retweet_num
```

```python
price = pd.read_csv('stock_price/CompanyValues.csv')
startPrice = '2015-01-05'
endPrice = '2019-12-27'
price = price[price['day_date'] >= startPrice]
price = price[price['day_date'] <= endPrice]
price_apple = price.copy()
price_google = price.copy()
price_amzn = price.copy()
price_tsla = price.copy()
price_msft = price.copy()
price_googl = price.copy()
price_apple = price_apple.loc[price['ticker_symbol'] == 'AAPL']
price_google = price_google.loc[price['ticker_symbol'] == 'GOOG']
price_googl = price_googl.loc[price['ticker_symbol'] == 'GOOGL']
price_amzn = price_amzn.loc[price['ticker_symbol'] == 'AMZN']
price_msft = price_msft.loc[price['ticker_symbol'] == 'MSFT']
price_tsla = price_tsla.loc[price['ticker_symbol'] == 'TSLA']
```

```python
#apple net price 
price_apple['day_date'] = pd.to_datetime(price_apple['day_date'], format="%Y/%m/%d")
logic = {'open_value'  : 'first',
         'close_value' : 'last',
        }
price_apple['day_date'] = pd.to_datetime(price_apple['day_date'])
price_apple = price_apple.set_index('day_date')
price_apple = price_apple.sort_index()
price_apple = price_apple.resample('W').apply(logic)
price_apple.index = price_apple.index - pd.tseries.frequencies.to_offset("7D")
price_apple['net_price'] = ((price_apple['close_value'] - price_apple['open_value']) / price_apple['open_value']) * 100
price_apple = price_apple.iloc[1:, :]
price_apple = price_apple.reset_index()
price_apple
```
